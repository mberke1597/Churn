```python
# Outlier Detection using IQR Method
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return len(outliers)

# Numeric columns to check for outliers
numeric_cols = ['CreditScore', 'Age', 'Balance', 'EstimatedSalary', 'Point Earned']

print("Outlier Analysis (IQR Method):")
print("=" * 50)
total_outliers = 0
for col in numeric_cols:
    outlier_count = detect_outliers(df, col)
    total_outliers += outlier_count
    outlier_percentage = (outlier_count / len(df)) * 100
    print(f"{col:20} : {outlier_count:5} outliers ({outlier_percentage:.2f}%)")

print("=" * 50)
print(f"{'Total Outliers':20} : {total_outliers:5}")
print(f"{'Total Rows':20} : {len(df):5}")
print(f"{'Outlier Ratio':20} : {(total_outliers / (len(df) * len(numeric_cols))) * 100:.2f}%")
```

```python
df = df.drop(['RowNumber', 'CustomerId', 'Surname'], axis = 1)
```

```python
hot = pd.get_dummies(df[['Geography', 'Gender', 'Card Type']])
```

```python
df = pd.concat([df, hot], axis = 1)
df = df.drop(['Geography', 'Gender', 'Card Type'], axis = 1)
```

```python
X = df.drop('Exited', axis = 1)
X = X.values
y = df['Exited']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_standard = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train, y_test = train_test_split(X_standard, y, test_size = 0.3, random_state = 0)
```

```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=0)
X_train, y_train = ros.fit_resample(X_train, y_train)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import ConfusionMatrixDisplay

naive_bayes = GaussianNB()
naive_bayes.fit(X_train, y_train)

previsoes = naive_bayes.predict(X_test)

ConfusionMatrixDisplay.from_predictions(y_test, previsoes)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report, ConfusionMatrixDisplay

naive_bayes = GaussianNB()
naive_bayes.fit(X_train, y_train)

previsoes = naive_bayes.predict(X_test)

print(classification_report(y_test, previsoes))

ConfusionMatrixDisplay.from_predictions(y_test, previsoes)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import RandomizedSearchCV

# Hiperparametre grid'i
parameters = {
    'max_depth': [3, 4, 5, 6, 7, 9, 11],
    'min_samples_split': [2, 3, 4, 5, 6, 7],
    'criterion': ['entropy', 'gini']
}

# Model
model = DecisionTreeClassifier()

# Random Search
grid_dt = RandomizedSearchCV(
    estimator=model,
    param_distributions=parameters,
    cv=3,
    n_jobs=-1,
    random_state=42
)

# Eitme
grid_dt.fit(X_train, y_train)

# Sonular
best_model = grid_dt.best_estimator_

print(" **Best Decision Tree Parameters**")
print(f" Min Samples Split : {best_model.min_samples_split}")
print(f" Max Depth         : {best_model.max_depth}")
print(f" Criterion         : {best_model.criterion}")
print(f" Best CV Score     : {grid_dt.best_score_:.4f}")
```
